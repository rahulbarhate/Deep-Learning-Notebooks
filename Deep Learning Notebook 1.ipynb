```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets.samples_generator import make_circles
```

```python
X, y = make_circles(n_samples = 1000,
                  noise = 0.1,
                  factor = 0.2,
                  random_state = 0)
```

```python
X
```

 X has 1000 rows (no. of samples) and 2 columns (features)

```python
X.shape
```

#### Plotting X and y values 

```python
plt.figure(figsize=(5, 5))
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5) #represented using a circle with blue colour for the values of x with y = 0
plt.plot(X[y==1, 0], X[y==1,1], 'xr', alpha=0.5) #represented using a cross with red colour for the values of x with y = 1
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.legend(['0', '1'])
plt.title("Blue circles and Red Crosses")
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
```

```python
model = Sequential()
```

```python
model.add(Dense(4, input_shape = (2,), activation = 'tanh'))
```

```python
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.compile(SGD(lr=0.5), 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(X, y, epochs = 20)
```

#### Drawing the decision boundary of our model

```python
# To do this we define a grid of points using a function from numpy called 'linspace'
# linspace - generates a sequence of numbers between a minimum value and the maximum value

hticks = np.linspace(-1.5, 1.5, 101) # params - (minimum, maximum, no of points)
vticks = np.linspace(-1.5, 1.5, 101)
aa, bb = np.meshgrid(hticks, vticks)
ab = np.c_[aa.ravel(), bb.ravel()]
c = model.predict(ab)
cc = c.reshape(aa.shape)
```

#### Plotting the data and overlay a contour plot of our prediction grid

```python
plt.figure(figsize=(6, 6))
plt.contourf(aa, bb, cc, cmap='bwr', alpha=0.2)
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5) 
plt.plot(X[y==1, 0], X[y==1,1], 'xr', alpha=0.5) 
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.legend(['0', '1'])
plt.title("Blue circles and Red Crosses")
```
